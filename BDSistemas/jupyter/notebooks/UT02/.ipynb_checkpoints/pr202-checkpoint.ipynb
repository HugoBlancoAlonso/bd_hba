```python
import redis

r = redis.Redis(
    host='redis',
    port=6379,
    db=0,
    decode_responses=True
)
```

```python
cliente = "Hugo"
producto = "Agua"

def agregar_pedido(cliente, producto):
    nuevoPedido = {
        "id": 1,
        "cliente": cliente,
        "producto": producto,
        "cantidad": 1,
        "urgente": False
    }

    r.rpush("pedidos", nuevoPedido)

agregar_pedido(cliente, producto)
```
