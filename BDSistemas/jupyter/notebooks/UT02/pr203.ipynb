```python
import redis
import json
```

```python
r = redis.Redis(
    host='redis',
    port=6379,
    db=0,
    decode_responses=True
)
```

```python
def addplayer(id, name, country, score):

    player_data = {
        "name": name,
        "country": country,
        "score": score
    }
    player_json = json.dumps(player_data)

    r.hset("player",str(id), player_json)
```

```python
addplayer(1, "hugo", "espa√±a", 1200)
```
