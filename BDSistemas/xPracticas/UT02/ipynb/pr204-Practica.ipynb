```python
def load_locations():
    for p in POIS:
        r.geoadd("poi:locations", (p["lon"], p["lat"], p["id"]))
        r.hset("poi:info", p["id"], p["name"])
    print("Datos guardados")
```

```python
load_locations()
```

```python
usuario = {
    "lat": 40.41677,
    "lon": -3.70379
}

def find_by_radius(lat,lon, distance=2000):
    locs = r.geosearch("poi:locations", latitude=lat, longitude=lon, radius=distance, unit="m")
    for loc in locs:
        name = r.hget("poi:info", loc)
        print(f" --> {name} ({loc})")
```

```python
find_by_radius(usuario["lat"], usuario["lon"])
```

```python
def find_nearest(lat, lon):
    loc = r.geosearch("poi:locations", latitude=lat, longitude=lon, radius=100000000000, count=1, withdist=True)
    nom = r.hget("poi:info", loc[0][0])
    print(f"El POI mas cercano es {nom}, a {loc[0][1]} KM")
```

```python
find_nearest(usuario["lat"], usuario["lon"])
```
