```python
import redis

r = redis.Redis(
    host='redis',
    port=6379,
    db=0,
    decode_responses=True
)
```

```python
actual_id = 1

def next_id():
    global actual_id
    id = "id_" + str(actual_id).zfill(4)
    actual_id = actual_id + 1
    return id
```

```python
import json

def agregar_pedido(cliente, producto, urg):
    if urg == "True":
        p = {
        "id": next_id(),
        "cliente": cliente,
        "producto": producto,
        "cantidad": 1,
        "urgente": True
        }
        productoJson = json.dumps(p)
        r.lpush("Pedidos", productoJson)
    else:
        p = {
        "id": next_id(),
        "cliente": cliente,
        "producto": producto,
        "cantidad": 1,
        "urgente": False
        }
        productoJson = json.dumps(p)
        r.rpush("Pedidos", productoJson)
```

```python
agregar_pedido("Hugo", "Agua")
```

```python
def procesar_pedido():
    p = r.lpop("Pedidos")
    pedido = json.loads(p)
    return pedido
```

```python
print(procesar_pedido())
```

```python
for i in range(5):
    nombre = input("Escribe tu nombre: ")
    pedido = input("Pedido: ")
    urg = input("Es urgente?: ")
    agregar_pedido(nombre, pedido, urg)
```

```python
todos = r.lrange("Pedidos", 0, -1)
for t in todos:
    t = json.loads(t)
    print(t)
```

```python
agregar_pedido("Josemi", "Bicarbonato")
agregar_pedido("Alfredo", "Pastis")
```

```python
total = r.llen("Pedidos")
for i in range(total):
    print(procesar_pedido())
```
